```python
def find_hashtags(text):
  hashtags = re.findall(r"#(\w+)", text)
  return hashtags

all_hashtags = []
for text in text_column:
  hashtags = find_hashtags(str(text))
  all_hashtags.extend(hashtags)

print(all_hashtags)
```

```python
df['hashtags'] = df["text"].apply(find_hashtags)
```

```python
def find_emails(text):
  emails = re.findall(r"[a-z0-9\.\-+_]+@[a-z0-9\.\-+_]+\.[a-z]+", text)
  return emails

all_emails = []
for text in text_column:
  emails = find_emails(str(text))
  all_emails.extend(emails)

print(all_emails)
```

```python
df['emails'] = df["text"].apply(find_emails)
```

```python
def remove_usernames(text):
  #return re.sub(r"@\w+", "", text)
  return re.sub(r'@\S+', '', text)

text_without_usernames = []
for text in text_column:
  text_without_usernames.append(remove_usernames(str(text)))

#print(text_without_usernames)
print(text_column[12])
print(text_without_usernames[12])
```

```python
def remove_links(text):
  return re.sub(r"https?://\S+|www\.\S+", "", text)

text_without_links = []
for text in text_column:
  text_without_links.append(remove_links(str(text)))

print(text_without_links)
```

```python
def remove_non_ascii_symbols(text):
  return "".join([char for char in text if ord(char) < 128])

text_without_non_ascii = []
for text in text_column:
  text_without_non_ascii.append(remove_non_ascii_symbols(str(text)))

print(text_without_non_ascii)
```

```python
def to_lower(text):
  return text.lower()

text_lower = []
for text in text_column:
  text_lower.append(to_lower(str(text)))

print(text_lower)
```

```python
def remove_stop_words(text):
  stop_words = set(stopwords.words('english'))
  words = text.split()
  filtered_words = [word for word in words if word.lower() not in stop_words]
  return " ".join(filtered_words)

text_without_stop_words = []
for text in text_column:
  text_without_stop_words.append(remove_stop_words(str(text)))

print(text_without_stop_words)
```

```python
def remove_digits(text):
  return re.sub(r'\d+', '', text)

text_without_digits = []
for text in text_column:
  text_without_digits.append(remove_digits(str(text)))

print(text_without_digits)
```

```python
def remove_special_characters(text):
  characters_to_remove = ['.', ',', '!', '?', '#', '$', '%', '&', '*', '(', ')',
                        '+', '=', '-', '_', '[', ']', '{', '}', ';', ':', '\'',
                        '"', '/', '\\', '|', '<', '>', '`', '~']
  for char in characters_to_remove:
    text = text.replace(char, '')
  return text




text_without_specific_characters = []
for text in text_column:
  text_without_specific_characters.append(
      remove_special_characters(str(text)))

print(text_without_specific_characters)
```

```python
df['new_text'] = df["text"].apply(remove_usernames)
df['new_text'] = df["new_text"].apply(remove_links)
df['new_text'] = df["new_text"].apply(remove_non_ascii_symbols)
df['new_text'] = df["new_text"].apply(to_lower)
df['new_text'] = df["new_text"].apply(remove_stop_words)
df['new_text'] = df["new_text"].apply(remove_digits)
df['new_text'] = df["new_text"].apply(remove_special_characters)
```
