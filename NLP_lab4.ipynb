```python
def to_lower(text):
  return text.lower()
```

```python
def remove_stop_words(text):
  stop_words = set(stopwords.words('english'))
  words = text.split()
  filtered_words = [word for word in words if word.lower() not in stop_words]
  return " ".join(filtered_words)
```

```python
def remove_non_ascii_symbols(text):
  return "".join([char for char in text if ord(char) < 123 and ord(char)>96 or ord(char) == 32])

text_without_non_ascii = []
for text in messages:
  text_without_non_ascii.append(remove_non_ascii_symbols(str(text)))
```

```python
new_text_train = []
for text in train_messages:
  new_text_train.append(remove_stop_words(remove_non_ascii_symbols(to_lower(str(text)))))

new_text_test = []
for text in test_messages:
  new_text_test.append(remove_stop_words(remove_non_ascii_symbols(to_lower(str(text)))))

print(new_text_train)
print(new_text_test)
```

```python
tokenized_text_train = []
for text in new_text_train:
  tokens =  nltk.word_tokenize(text)
  tokenized_text_train.append(tokens)

tokenized_text_test = []
for text in new_text_test:
  tokens =  nltk.word_tokenize(text)
  tokenized_text_test.append(tokens)

print(tokenized_text_train)
print(tokenized_text_test)
```

```python
def create_vocabulary(tokenized_text):
  vocabulary = set()
  for tokens in tokenized_text:
    for token in tokens:
      vocabulary.add(token)
  return vocabulary

#print(create_vocabulary(tokenized_text))
vacabluary = create_vocabulary(tokenized_text_train)
#vacabluary_test = create_vocabulary(tokenized_text_test)

print(vacabluary)
#print(vacabluary_test)
```

```python
def document_to_bow(tokens, vocabulary):
    bow_vector = np.zeros(len(vocabulary))
    vocabulary_list = list(vocabulary) #- convert the set to a list to allow for indexing
    vocabulary_dict = {token: index for index, token in enumerate(vocabulary_list)} # build vocabulary dictionary from list
    for token in tokens:
        if token in vocabulary_dict:
            bow_vector[vocabulary_dict[token]] += 1 # access the index using the dictionary
    return bow_vector

# Use the 'vacabluary' variable created in the previous cell and rename it to vocabulary
vocabulary_train = vacabluary # Assign the vacabluary to vocabulary_train
# Create vocabulary for test data
vocabulary_test = create_vocabulary(tokenized_text_test)

bow_train = []
for document_tokens in tokenized_text_train:  # Iterate through each document's tokens
    bow_vector = document_to_bow(document_tokens, vocabulary_train) # Use vocabulary_train instead of vacabluary_train
    bow_train.append(bow_vector)  # Append the bow vector to the list


bow_test = []
for document_tokens in tokenized_text_test:
    bow_vector = document_to_bow(document_tokens, vocabulary_test) # Use vocabulary_test instead of vacabluary_test
    bow_test.append(bow_vector)

print(bow_train[0])
print(bow_test[0])
```

```python
def train_naive_bayes(tokenized_messages, labels, vocabulary):
    spam_count = sum(labels == 'spam')
    ham_count = sum(labels == 'ham')
    total_count = len(labels)

    prior_spam = spam_count / total_count
    prior_ham = ham_count / total_count

    spam_word_count = {word: 0 for word in vocabulary}
    ham_word_count = {word: 0 for word in vocabulary}

    for i, tokens in enumerate(tokenized_messages):
        for token in tokens:
            if labels.iloc[i] == 'spam':
                spam_word_count[token] += 1
            else:
                ham_word_count[token] += 1

    total_spam_words = sum(spam_word_count.values()) + len(vocabulary)
    total_ham_words = sum(ham_word_count.values()) + len(vocabulary)

    for word in vocabulary:
        spam_word_count[word] = (spam_word_count[word] + 1) / total_spam_words
        ham_word_count[word] = (ham_word_count[word] + 1) / total_ham_words

    return prior_spam, prior_ham, spam_word_count, ham_word_count
```

```python
def classify_message(message, prior_spam, prior_ham, spam_word_count, ham_word_count, vocabulary):
    tokens = preprocess_text(message)

    log_prob_spam = np.log(prior_spam)
    log_prob_ham = np.log(prior_ham)

    for token in tokens:
        if token in vocabulary:
            log_prob_spam += np.log(spam_word_count[token])
            log_prob_ham += np.log(ham_word_count[token])

    return 'spam' if log_prob_spam > log_prob_ham else 'ham'
```

```python
def evaluate_model(test_labels, predictions):
    correct = sum(test_labels.values == predictions)
    accuracy = correct / len(test_labels)
    return accuracy
```

```python
def preprocess_text(text):
    text = to_lower(text)
    text = remove_non_ascii_symbols(text)
    text = remove_stop_words(text)
    tokens = nltk.word_tokenize(text)
    return tokens

prior_spam, prior_ham, spam_word_count, ham_word_count = train_naive_bayes(tokenized_text_train, train_labels, vocabulary_train)

predictions = []
for message in test_messages:
  prediction = classify_message(message, prior_spam, prior_ham, spam_word_count, ham_word_count, vocabulary_train)
  predictions.append(prediction)

accuracy = evaluate_model(test_labels, predictions)
print("Accuracy:", accuracy)
```
